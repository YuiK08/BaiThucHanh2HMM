```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Bước 1: Lấy dữ liệu quan sát từ cảm biến ===
# Mỗi hàng là (ax, ay): gia tốc theo trục x và y
observations = df_after[['userAcceleration.x', 'userAcceleration.y']].to_numpy()

# === Bước 2: Cấu hình hệ thống Particle Filter ===
NUM_PARTICLES = 500     # Số lượng hạt
DT = 0.1                # Bước thời gian
NOISE_STD_POS = 0.2     # Nhiễu vị trí
NOISE_STD_VEL = 0.1     # Nhiễu vận tốc
OBS_NOISE_STD = 0.5     # Nhiễu quan sát (gia tốc)

# === Bước 3: Khởi tạo trạng thái ẩn cho tất cả hạt ===
particles = {
    'x': np.random.normal(0, 1, NUM_PARTICLES),    # vị trí x
    'y': np.random.normal(0, 1, NUM_PARTICLES),    # vị trí y
    'vx': np.zeros(NUM_PARTICLES),                 # vận tốc x
    'vy': np.zeros(NUM_PARTICLES),                 # vận tốc y
    'w': np.ones(NUM_PARTICLES) / NUM_PARTICLES    # trọng số khởi tạo
}

# === Bước 4: Mô hình chuyển trạng thái (Transition Model) ===
def transition_model(particles, ax, ay):
    # Dự đoán vị trí mới
    particles['x'] += particles['vx'] * DT + np.random.normal(0, NOISE_STD_POS, NUM_PARTICLES)
    particles['y'] += particles['vy'] * DT + np.random.normal(0, NOISE_STD_POS, NUM_PARTICLES)

    # Dự đoán vận tốc mới từ gia tốc đo được
    particles['vx'] += ax * DT + np.random.normal(0, NOISE_STD_VEL, NUM_PARTICLES)
    particles['vy'] += ay * DT + np.random.normal(0, NOISE_STD_VEL, NUM_PARTICLES)

    return particles

# === Bước 5: Mô hình quan sát (Observation Model) ===
def observation_model(particles, ax, ay):
    # Tính khoảng cách giữa vận tốc của hạt và gia tốc đo được
    dist = np.linalg.norm([particles['vx'] - ax, particles['vy'] - ay], axis=0)
    weights = np.exp(- (dist ** 2) / (2 * OBS_NOISE_STD ** 2))
    return weights + 1e-300  # tránh chia cho 0

# === Bước 6: Tái lấy mẫu (Resampling) ===
def resample(particles):
    idx = np.random.choice(np.arange(NUM_PARTICLES), size=NUM_PARTICLES, p=particles['w'])
    for key in ['x', 'y', 'vx', 'vy']:
        particles[key] = particles[key][idx]
    particles['w'] = np.ones(NUM_PARTICLES) / NUM_PARTICLES  # reset trọng số
    return particles

# === Bước 7: Vòng lặp Particle Filtering ===
estimated_positions = []

for ax, ay in observations:
    # (1) Dự đoán trạng thái mới theo mô hình chuyển trạng thái
    particles = transition_model(particles, ax, ay)

    # (2) Cập nhật trọng số từ mô hình quan sát
    weights = observation_model(particles, ax, ay)
    particles['w'] = weights / np.sum(weights)

    # (3) Ước lượng trạng thái trung bình trọng số
    x_est = np.average(particles['x'], weights=particles['w'])
    y_est = np.average(particles['y'], weights=particles['w'])
    estimated_positions.append([x_est, y_est])

    # (4) Tái lấy mẫu các hạt theo trọng số mới
    particles = resample(particles)


trajectory_df = pd.DataFrame(estimated_positions, columns=["x", "y"])

plt.figure(figsize=(10, 6))
plt.plot(trajectory_df['x'], trajectory_df['y'], label='Quỹ đạo ước lượng', color='blue')
plt.title('Hiển thị quỹ đạo ước lượng Particle Filter (Nonlinear HMM)')
plt.xlabel('Position X')
plt.ylabel('Position Y')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()

# (Tùy chọn) Lưu kết quả ra CSV
trajectory_df.to_csv("estimated_trajectory.csv", index=False)
```
